```python
# Colby Whitehouse
# 12/8/2023
# graphData
# This program takes a .csv file with latitude, longitude, accuracy, and time and creates 3 plots for the data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import csv

# Let the user upload files
uploaded = files.upload()
uploaded = files.upload()
file_name = 'SDHdata.csv'
file_place_name = 'SDHdataPlaces.csv'

# Open and read the first .csv file
with open(file_name, newline ='') as csvfile:
  data = list(csv.reader(csvfile))
print(data)

# Open and read the second .csv file
with open(file_place_name, newline ='') as csvfile:
  places = list(csv.reader(csvfile))
print(places)


# Plotting the values versus time
fig = plt.figure(figsize=(15, 6))

latitude = []
longitude = []
accuracy = []
time = []
place = []
place_time = []
n = 0

# Place the data in arrays to plot
for entry in data:
  latitude.append(float(entry[0]))
  longitude.append(float(entry[1]))
  accuracy.append(float(entry[2]))
  time.append(entry[3])
  n+=1

# Plot the data in 3 subplots
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax.plot(time, latitude, 'k-o', label = 'Latitude')
ax.legend()
ax.set_xticks(time[::25])
ax.set_title("Location Data SDH")

ax2.plot(time, longitude, 'k-+', label = 'Longitude')
ax2.legend()
ax2.set_xticks(time[::25])
ax3.plot(time, accuracy, 'k-*', label = 'Accuracy')
ax3.legend()
ax3.set_xticks(time[::25])



# Show the plot
plt.show()

# Calculate and print mean and standard deviation of the accuracy
mean_accuracy = np.mean(accuracy)
std_accuracy = np.std(accuracy)

print("The mean accuracy is: " + str(mean_accuracy))
print("The strandard deviation of accuracy is: " + str(std_accuracy))
```
